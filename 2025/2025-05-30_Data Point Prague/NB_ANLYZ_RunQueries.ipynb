```python
# Set the bases

dataset_name = "Sales Analysis"
```

```python
import sempy.fabric as fabric
```

```python
# Get entire table
df_table = fabric.read_table(dataset_name, "date")
df_table.head(10) #show only first 10 rows
```

```python
# Query semantic model and return table
fabric.evaluate_measure(dataset_name, \
    measure=["# Sales"], \
    groupby_columns=["product[Color]"], \
    filters={"product[Color]": ["Black", "Blue", "Yellow"]})
```

```python
## Run query fetched from Perf. Analyzer in PBI Desktop

(fabric.evaluate_dax(
        dataset = dataset_name,
        dax_string=
        """
// DAX Query
DEFINE
	VAR __DS0Core = 
		SUMMARIZECOLUMNS('date'[FiscalYear], 'date'[FiscalQuarter], "v__Sales", 'sales'[# Sales])

	VAR __DS0PrimaryWindowed = 
		TOPN(1001, __DS0Core, 'date'[FiscalYear], 1, 'date'[FiscalQuarter], 1)

EVALUATE
	__DS0PrimaryWindowed

ORDER BY
	'date'[FiscalYear], 'date'[FiscalQuarter]

        """  
       )  
)
```
