## Measure table
Given semantic models can only be created containing at least 1 table, we must have a placeholder table. We create the measure table here, which is a good practice. 

The name "Measures" is reserved. 

```python
# Create empty dataframe as measure table

from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType

# Define a schema with one column
schema = StructType([
    StructField("Message", StringType(), True)
])

# Create a DataFrame with one row and one column
data = [(" ",)]
df = spark.createDataFrame(data, schema=schema)

# Show the DataFrame (just to verify)
df.show()

# Save to LH_STORE_Gold (your default Lakehouse) as a new table
df.write.mode("overwrite").saveAsTable("_Measures")
```

## Create semantic model

```python
labs.directlake.generate_direct_lake_semantic_model(
    dataset= semanticmodel_name, 
    lakehouse_tables= '_measures', 
    workspace= None, 
    lakehouse= None, # defaults to the attached lakehouse
    lakehouse_workspace= None, 
    schema= 'dbo', 
    overwrite = True, 
    refresh= False)
```
