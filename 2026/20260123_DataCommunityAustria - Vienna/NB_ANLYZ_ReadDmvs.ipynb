# Semantic-link basic setup

```python
# Set the bases

dataset_name = "Example Semantic Model"
lakehouse_name = "SempyDemo"
```

```python
import sempy.fabric as fabric
# from sempy_relationships import plot_relationship_metadata
```

# Get Semantic Table info
- Get Tables


```python
# Show model tables through Sempy function
tables = fabric.list_tables(dataset_name)
tables
```

```python
## Get tables properties through DMV

(fabric
    .evaluate_dax(
        dataset = dataset_name,
        dax_string=
        """
        select * from $SYSTEM.TMSCHEMA_TABLES
        """  
       )  
)
```

# Get Measure information

```python
## Get measure properties through DMV

(fabric
    .evaluate_dax(
        dataset = dataset_name,
        dax_string=
        """
        SELECT * FROM $SYSTEM.TMSCHEMA_MEASURES
        """  
       )  
)
```

# Semantic Model Relationships
