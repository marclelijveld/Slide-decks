```python
# Set the bases

dataset_name = "Example Semantic Model"
lakehouse_name = "SempyDemo"
```

```python
import sempy.fabric as fabric
```

```python
# Show entire table
df_table = fabric.read_table(dataset_name, "DimProduct")
df_table
```

```python
# Query semantic model and return table
fabric.evaluate_measure(dataset_name, \
    measure=["$ Sales", "# Orders"], \
    groupby_columns=["DimProduct[Color]"], \
    filters={"DimProduct[Color]": ["Black", "Blue", "Yellow"]})
```

```python
## Run query fetched from Perf. Analyzer in PBI Desktop

(fabric.evaluate_dax(
        dataset = dataset_name,
        dax_string=
        """
// DAX Query
DEFINE
	VAR __DS0FilterTable = 
		TREATAS({2020}, 'DimDate'[FiscalYear])

	VAR __DS0Core = 
		SUMMARIZECOLUMNS(
			'DimProductSubcategory'[Sub Category Name],
			__DS0FilterTable,
			"v__Orders", 'FactInternetSales'[# Orders]
		)

	VAR __DS0PrimaryWindowed = 
		TOPN(1001, __DS0Core, [v__Orders], 0, 'DimProductSubcategory'[Sub Category Name], 1)

EVALUATE
	__DS0PrimaryWindowed

ORDER BY
	[v__Orders] DESC, 'DimProductSubcategory'[Sub Category Name]


        """  
       )  
)
```
