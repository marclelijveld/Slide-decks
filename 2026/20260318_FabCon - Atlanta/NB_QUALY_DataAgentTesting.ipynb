## Fixed question / answer testing
This setup uses a fixed question and expected answer to validate through the Data Agent. All questions and answers are hard-coded in this example.

## Dynamic testing 
This setup uses dynamic questions and expected answers. As data changes over time, the results to validate might have changed as well. 
In this setup, we use a metadata setup where we define the prompt, source that contains the answer, and the query that should return the answer. After which the query is directly executed on the source and answer will be compared with running the prompt through the data agent. 

### Generate dynamic expected answers

```python
# Get evaluation set
df_testset = spark.sql(f"SELECT * FROM {lakehouse_name}.dbo.dynamic_evaluation_prompts")
display(df_testset)
```

```python
# Get evaluation set
df_testset = spark.sql(f"SELECT * FROM {lakehouse_name}.dbo.dynamic_evaluation_prompts")

# Convert to pandas for iteration
rows = df_testset.toPandas()

results = []

for _, row in rows.iterrows():

    source_type = row.SourceType
    prompt = row.Prompt
    query = row.Query

    print(f"Processing Prompt: '{prompt}'")
    print(f"SourceType: '{source_type}'")
    print(f"Query preview: {query[:200]}...\n")

    # -----------------------------------------
    # Semantic Model execution
    # -----------------------------------------
    if source_type == "Semantic Model":

        workspace_name = row.WorkspaceName
        dataset_name = row.ExpectedSource

        print(f"Workspace: '{workspace_name}'")
        print(f"Semantic Model: '{dataset_name}'")

        result = fabric.evaluate_dax(
            workspace=workspace_name,
            dataset=dataset_name,
            dax_string=query
        )

        results.append({
            "Prompt": prompt,
            "SourceType": source_type,
            "Source": dataset_name,
            "ExecutedQuery": query,
            "Result": result
        })


    # -----------------------------------------
    # Lakehouse execution
    # -----------------------------------------
    elif source_type == "Lakehouse":

        lakehouse_id = row.WorkspaceName

        print(f"Lakehouse ID: '{lakehouse_id}'")

        df_result = spark.sql(query)
        result_pd = df_result.toPandas()

        results.append({
            "Prompt": prompt,
            "SourceType": source_type,
            "Source": lakehouse_id,
            "ExecutedQuery": query,
            "Result": result_pd
        })


# Convert combined results to dataframe
results_df = pd.DataFrame(results)

display(results_df)
```

```python
df_dynamictestset = pd.DataFrame(
    columns=["question", "expected_answer"],
    data=[
        [row["Prompt"], str(row["Result"].iloc[0, 0]) if hasattr(row["Result"], "iloc") else str(row["Result"])]
        for _, row in results_df.iterrows()
    ]
)

display(df_dynamictestset)
```

#### Start dynamic test through SDK

```python
# Two tables will be created:
# - "<table_name>": contains summary results (e.g., accuracy)
# - "<table_name>_steps": contains detailed reasoning and step-by-step execution
table_name = "dynamic_test_output"

# Specify the Data Agent stage: "production" (default) or "sandbox"
data_agent_stage = "production"

# Run the evaluation and get the evaluation ID
evaluation_id = evaluate_data_agent(
    df_dynamictestset,
    data_agent_name,
    workspace_name=workspace_name,
    table_name=table_name,
    data_agent_stage=data_agent_stage
)

print(f"Unique ID for the current evaluation run: {evaluation_id}")
```
